1.  What is the difference between a function and a method in Python?
     - Function: A function is a block of reusable code that performs a specific task. It is defined using the def keyword and can be used independently.

     def greet():
     
     print("Hello!")
     
     greet()
     
     Method: A method is a function that is associated with an object. It is called using the dot (.) operator on objects (usually of a class or data type like string or list).

     name = "Alice"
   
     print(name.upper())  # 'upper' is a method of string object

5. What are iterators in Python and how do they differ from iterables?
     - Iterable: Any object capable of returning its members one at a time (like list, tuple, string).

   - Iterator: An object with a __next__() method that returns the next item.

     lst = [1, 2, 3]

     it = iter(lst)  # it is now an iterator

     print(next(it))  # Output: 1

6. Explain the concept of generators in Python and how they are defined.
     - Generators are iterators that yield values one at a time using the yield
       keyword.

       They are defined like functions, but use yield instead of return.

        def count_up_to(n):
    
          i = 1
    
          while i <= n:
        
          yield i
        
          i += 1


7. What are the advantages of using generators over regular functions?
     - Memory efficient: Generate items one at a time, not all at once.

     - Faster for large data: No need to store all values in memory.

     - Convenient for infinite or large sequences.



8. What is a lambda function in Python and when is it typically used?
     - A lambda function is a small anonymous function defined with the lambda
       keyword.

      Used for short operations or as an argument to functions like map(), filter().

      square = lambda x: x * x

      print(square(5))  # Output: 25

9. Explain the purpose and usage of the map() function in Python.
     - Applies a function to each item in an iterable and returns a map object (iterator).

      nums = [1, 2, 3]

      squared = map(lambda x: x**2, nums)

      print(list(squared))  # [1, 4, 9]

10. What is the difference between map(), reduce(), and filter() functions in Python?
     - map():	Applies a function to every element and returns	Iterator
     - filter():	Selects elements that meet a condition and returns Iterator
     - reduce():	Applies a function cumulatively to elements and returns	Single value

      Example:

      from functools import reduce

      nums = [1, 2, 3, 4]

      print(list(map(lambda x: x*2, nums)))       # [2, 4, 6, 8]

      print(list(filter(lambda x: x % 2 == 0, nums)))  # [2, 4]

      print(reduce(lambda x, y: x + y, nums))     # 10
